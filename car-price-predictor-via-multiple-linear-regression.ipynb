```python
# read dataset
dataset = pd.read_csv('/kaggle/input/car-prices-prediction-data/CarPricesPrediction.csv')

# drop 1st column, row count
dataset = dataset.drop(dataset.columns[0], axis=1)
dataset = dataset.drop('Model', axis=1)

X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

make_cond = ['Make', 'Condition']

ct = ColumnTransformer(transformers = [('encoder', OneHotEncoder(), [0,3])], remainder = 'passthrough')

X = np.array(ct.fit_transform(X))

#y = dataset['Price']
```

# **Train Test Split**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

# **Train the Model**

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

# Execute the Prediction

```python
y_pred = regressor.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

# Determine the Accuracy

```python
#𝑅^2 : R squared is another way to evaluate the performance of a regression model.
# 1, means that the model is perfect and 0 means the the model will perform poorly.
r2_reg_model_test = regressor.score(X_test, y_test)

print("R^2 Test: {}".format(r2_reg_model_test))
```

```python
from sklearn.metrics import mean_absolute_error
print("MAE",mean_absolute_error(y_test,y_pred))
```
